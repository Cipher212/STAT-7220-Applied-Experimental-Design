**Question 7**

```python
# ANOVA

model = ols('Comfort ~ C(Material)', data=df).fit()


residuals = model.resid

anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:")
print(anova_table)

print("\n\n")

# Shapiro Wilke
shapiro_test = stats.shapiro(residuals)
print("Shapiro-Wilk Test for Normality (Residuals):")
print(f"Test Statistic (W): {shapiro_test.statistic:.4f}")
print(f"p-value: {shapiro_test.pvalue:.4f}\n")

# qqplot

plt.figure(figsize=(7, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of ANOVA Residuals', fontsize=14)
plt.xlabel('Theoretical Quantiles', fontsize=12)
plt.ylabel('Standardized Residuals', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```


Both methods support the assumption of normality: the Shapiro-Wilk p-value (0.3874) is greater than 0.05, and the Q-Q plot residuals closely track the expected diagonal line, considering that we are dealing with a low sample size. Although there is some fall off at the tails.

**Question 8**
Homogenity of Variance


```python
# 4 groups
plastic = df[df['Material'] == 'Plastic']['Comfort']
rubber = df[df['Material'] == 'Rubber']['Comfort']
silicone = df[df['Material'] == 'Silicone']['Comfort']
metal = df[df['Material'] == 'Metal']['Comfort']

#leven's test

levene_stat, levene_p = stats.levene(plastic, rubber, silicone, metal)
print("Levene's Test for Homogeneity of Variance:")
print(f"Test Statistic (W): {levene_stat:.4f}")
print(f"p-value: {levene_p:.4f}\n")


model = ols('Comfort ~ C(Material)', data=df).fit()

plt.figure(figsize=(7, 6))


plt.scatter(model.fittedvalues, model.resid, color='blue', edgecolors='k', alpha=0.7, s=60)
plt.axhline(0, color='red', linestyle='--') # Add a reference line at zero
plt.title('Residuals vs. Fitted Values', fontsize=14)
plt.xlabel('Fitted Values (Predicted Group Means)', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```

Levene's P-Test W= ~ 0.579 with p-value > alpha shows that assumption of equal variance is met.

There is no distinct patterning or funnneling in the fitted vs residuals plot either. Points are mostly uniform in their spread for all 4 material types.
