**Question 5**
**Ans:**

**Null Hypothesis (H0):** Mean perceived comfort level is equal for all 4 grip materials

**Alternate Hypothesis (Ha):** At least one grip material has a mean perceived comfort level that is statistically significant in its from difference from the others

**Question 6**
**Ans:**

```python

df=pd.read_excel("Video Game Controllers.xlsx")

# EDA

#summary
summary_stats = df.groupby('Material')['Comfort'].describe()
print("Summary")
print(summary_stats)

print("\n\n\n")

#boxplot
sns.boxplot(x='Material', y='Comfort', hue='Material', data=df, palette='Set3', legend=True)



plt.title('Comfort Ratings by Grip Material', fontsize=14)
plt.ylabel('Comfort Rating', fontsize=17)
plt.xlabel('Material', fontsize=15)
plt.ylim(3, 9)



#density plot

plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Comfort', hue='Material', fill=True, palette='Set2', alpha=0.5)
plt.title('Density Plot of Comfort Ratings by Material')
plt.xlabel('Comfort Rating')
plt.ylabel('Density')
plt.savefig('comfort_density.png', bbox_inches='tight')
```

**Observations:** ANOVA test assumes normal distribution and equal variances which is mostly met by the scattering of observations for the 4 materials, despite having such a small sample size. Few outliers for Metal and Silicone cause it to appear almost bimodial.

Silicone appears to have the highest comfort ratings at ~ 7.6 CR

**Question 7**

```python
# ANOVA

model = ols('Comfort ~ C(Material)', data=df).fit()


residuals = model.resid

anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:")
print(anova_table)

print("\n\n")

# Shapiro Wilke
shapiro_test = stats.shapiro(residuals)
print("Shapiro-Wilk Test for Normality (Residuals):")
print(f"Test Statistic (W): {shapiro_test.statistic:.4f}")
print(f"p-value: {shapiro_test.pvalue:.4f}\n")

# qqplot

plt.figure(figsize=(7, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of ANOVA Residuals', fontsize=14)
plt.xlabel('Theoretical Quantiles', fontsize=12)
plt.ylabel('Standardized Residuals', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
```


Both methods support the assumption of normality: the Shapiro-Wilk p-value (0.3874) is greater than 0.05, and the Q-Q plot residuals roughly follow the expected diagonal line, considering that we are dealing with a low sample size. Although there is some fall off at the tails.

**Question 8**
Homogenity of Variance


```python
# 4 groups
plastic = df[df['Material'] == 'Plastic']['Comfort']
rubber = df[df['Material'] == 'Rubber']['Comfort']
silicone = df[df['Material'] == 'Silicone']['Comfort']
metal = df[df['Material'] == 'Metal']['Comfort']

#leven's test

levene_stat, levene_p = stats.levene(plastic, rubber, silicone, metal)
print("Levene's Test")
print(f"test Statistic: {levene_stat:.3f}")
print(f"p-Value: {levene_p:.3f}")


model = ols('Comfort ~ C(Material)', data=df).fit()

plt.figure(figsize=(7, 6))


plt.scatter(model.fittedvalues, model.resid, color='black', alpha=0.7, s=60)
plt.axhline(0, color='red', linestyle='--') # ref line
plt.title('Residuals vs. Fitted Values', fontsize=14) # title
plt.xlabel('Fitted Values (Predicted Group Means)', fontsize=12) # xlabel
plt.ylabel('Residuals', fontsize=15) #ylabel
plt.grid(True, linestyle='-', alpha=0.6)
```

Levene's P-Test W= ~ 0.579 with p-value > alpha shows that assumption of equal variance is met.

There is no distinct patterning or funnneling in the fitted vs residuals plot either. Points are mostly uniform in their spread for all 4 material types.

**Question 9**

```python
anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:")
print(anova_table)
```

**Observation:**
From the ANOVA table, the F-stat of **17.07** and significantly small p-value strongly support the alternative hypothesis (Ha) of at least one material group having a statistically significant difference in mean perceived comfort compared to the rest.



**Question 10**

```python
tukey_results = pairwise_tukeyhsd(endog=df['Comfort'], groups=df['Material'], alpha=0.05)

print(tukey_results)
```

**Observation**: For silicone we see 3 'True' results with adjusted p-value < 0.05. Rest are not significantly different from each other. Fail to reject.



**Question 11**

Conclusion:

Analysis confirms what was implied in the EDA, that we can reject the null hypothesis and say that at least one grip mateiral (Silicone), was significantly different from the rest in terms of perceived comfort

Our tests confirmed that the data meets all assumptions (Normality, Equal variance) with Shaprio-Wilke (p= ~ 0.39) and Leven's P (p= ~ 0.63). ANOVA F-test also supported Ha hypothesis with F-score 17.07 and p-value < alpha. The Tukey's HSD test also showed statistic significant difference in perceived comfort for Silicone compared to rest. However remaining 3 materials did nt show significant difference form each other.

Limitations:

- Scope of experiment was too limited. Style of gameplay of Apex Legends is not necessarily same for other genres. Grip styles and intensity of use would be more varied

- Using only pro players means that the controller grip is not accounting for casual player's taste which will be 99% of the consumer base

- It is not clear what the duration of testing was. Although one month is mentioned, it is not clear if every player used their controller for equal duration during that time period.  Many controller grips are fine when new, but wear down quickly and become sticky. This was not accounted for in data collection


Improvements:

If this study is repeated, game controller company should recruit a mix of casual and professional gamers across various game genres. Additionally, extending the trial period to 3–6 months or adding a secondary rating scale specifically for resistance to wear would provide a much more realistic picture of each material's viability in the market.

**Question 7**


```python
# ANOVA
model2 = ols('Score ~ C(Method) * C(Experience)', data=df2).fit()

residuals2 = model2.resid

anova_table2 = sm.stats.anova_lm(model2, typ=2)
print("Two-Way ANOVA Table:")
print(anova_table2)

print("\n")

# Shapiro Wik test

shapiro_test2 = stats.shapiro(residuals2)
print("Shapiro-Wilk Test:")
print(f"W Statistic): {shapiro_test2.statistic:.3f}")
print(f"p-value: {shapiro_test2.pvalue:.3f}")


# QQplot

plt.figure(figsize=(7, 6))
stats.probplot(residuals2, dist="norm", plot=plt)
plt.title(' qqPlot - 2 Way ANOVA', fontsize=14)
plt.xlabel('Theoretical', fontsize=12) # xlabel
plt.ylabel('Standardized Residuals', fontsize=12) # yalabel
plt.grid(True, linestyle='-', alpha=0.6) # solid gridline

plt.show()
```

**Observation:**

The visual of the QQplot and the Shapiro Wilke test both support the assumption of normality for the 2 Way ANOVA model.

```python

# divide into 4 groups based on Method and Experience combinations
g1 = df2[(df2['Method'] == 'Traditional') & (df2['Experience'] == 'No Prior Experience')]['Score']
g2 = df2[(df2['Method'] == 'Traditional') & (df2['Experience'] == 'Prior Experience')]['Score']
g3 = df2[(df2['Method'] == 'VR') & (df2['Experience'] == 'No Prior Experience')]['Score']
g4 = df2[(df2['Method'] == 'VR') & (df2['Experience'] == 'Prior Experience')]['Score']


# levene's test
levene_stat, levene_p = stats.levene(g1, g2, g3, g4)
print("Levene's Test")
print(f"Test Statistic (W): {levene_stat:.3f}")
print(f"p-value: {levene_p:.3f}")

print("\n") # blank


plt.figure(figsize=(7, 6))


# scatterplot
plt.scatter(model2.fittedvalues, model2.resid, color='yellow', edgecolors='b', alpha=0.7, s=60)
plt.axhline(0, color='red', linestyle='--') # reference line
plt.title('Residuals vs. Fitted', fontsize=14) # title
plt.xlabel('Fitted', fontsize=12) #xlabel
plt.ylabel('Residuals', fontsize=12) #yaleb
plt.grid(True,linestyle='-') # solid gridline
```

**Observation** Both the visual observstion of the plot and the Shapiro Wilke's test show that there is equal variance across the 4 groups are relatively equal.

Predicted score for Traditional - Experience (30.9) and Traditional - No Experience (30.7) is very similar. Fitted vs Residuals scatterplot shows no distinct funneling pattern. Distribution of points is mostly even across the 4 possible combinations of teaching methods and experienc levels.

The Levene's test has p-value > 0.05 (fail to reject Null) which in this case is homogenity of variance

**Question 1**
**Ans:** Human data consists of high variance due to individual differences (age, height, weight, medical condition, strength) and CRD essentially 'ignores' all these factors as it does not include any blocking factors. It will randomly distribute the varaince from these treatment effects.

For example, it would have ignored any prior experience for students in the Traditional or VR classes.


**Question 2**
**Ans:** Relying on ANOVA would only tell us that there exists a significant statistical difference for one group compared to the others in a particular dataset. However it would not tell us exactly which one. This is why we use post-hoc tests like TukeyHSD


**Question 3**
**Ans:**

RBD controls for one blocking factor whereas a Latin Square can do it for 2, and variations of LSD can do it for more than 2 blocking factors.


**Question 4**
**Ans:** Randomisation is important to prevent bias in experiments and to evenly distribute lurking variables that were not accounted for to all groups.


**Question 5**
**Ans:** If experiments are designed without replication in mind, as in, having only one sample per individual group in a block experiment, it would mean almost no low statistical power and precision. Inability to do basic tests like ANOVA as you would zero variation in a group with a singular sample.


On the other hand, Replication is also important in the design of experiments because collecting real world data costs time and money. A good expeirmental design  means deciding beforehand the MINIMUM samples required to have appropriate amount of statistical power through a power analysis.
