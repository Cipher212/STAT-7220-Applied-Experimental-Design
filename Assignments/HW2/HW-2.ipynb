**Question 7**

```python
model = ols('Score ~ C(Teaching_Method) * C(Prior_Knowledge)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
residuals = model.resid


#qq plot

fig, ax = plt.subplots(figsize=(6, 5))
sm.qqplot(residuals, line='s', ax=ax)
plt.title('Q-Q Plot of residuals')
plt.show()


# shapiro wilkes


# H0:  Dataset is not normally distributed
# HA: Dataset IS normally distributed

stat, p_value = shapiro(residuals)
print(f"Shapiro-Wilk Test Statistic: {stat:.4f}")
print(f"p-value: {p_value:.4f}")
```

*Interpretation*


The QQplot of residuals supports assumption of normality as it follows the line mostly, except at the tails. Shapiro Wilke Test Statistic of 0.9886 with p-value > 0.05 also strongly supports assumption of normality as Null hypothesis is rejected.
